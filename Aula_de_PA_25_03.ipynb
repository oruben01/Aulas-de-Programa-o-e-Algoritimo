```python
# Escrever um programa que recebe 04 notas do aluno, calcula a média e mostra se o aluno foi aprovado, recuperação ou reprovado.

# Entrada
print ("Digite suas notas AQUI, valores de 01 a 10")
nota1 = float (input("Digite a nota do 1º bimestre: "))
nota2 = float (input("Digite a nota do 2º bimestre: "))
nota3 = float (input("Digite a nota do 3º bimestre: "))
nota4 = float (input("Digite a nota do 4º bimestre: "))

# Processamneto

resultado = (nota1 + nota2 + nota3 + nota4) / 4

if resultado >= 7:

  print (f"A média das suas notas foi {resultado}, você foi aprovado. ")

elif resultado >= 5:

  print (f"A média da suas notas foi {resultado}, você terá que fazer recuperação")

else:

  print (f"A média das suas notas foi {resultado}, você foi reprovado.")

# Saída

```
