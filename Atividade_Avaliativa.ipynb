```python
# 1 - Par ou Ímpar

# Entrada

num =   int(input("Digite qualquer número: "))

# Processamento e Saída

if num % 2 == 0:
  print (f"O número {num}, é um número par. ")
else:
  print (f"O núemro {num}, é um número ímpar. ")
```

```python
# 2 - Maior de dois números

# Entrada

num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))

# Processamento e Saída

if num1 > num2:
  print(f"O número {num1} é MAIOR que o número {num2}. ")

elif num2 > num1:
  print (f"O número {num2} é MAIOR que o número {num1}. ")

else:
  print(f"O número {num1} e o número {num2} são IGUAIS.")
```

```python
# 3 - Aprovação de aluno
# Peça a nota de um aluno e informe:
# Aprovado (nota ≥ 7)
# Recuperação (5 ≤ nota < 7)
# Reprovado (nota < 5)

# Entrada

nome = str (input("Digite seu nome: "))
nota1 = float(input("Digite sua nota do 1º bimestre: "))
nota2 = float(input("Digite sua nota do 2º bimestre: "))
nota3 = float(input("Digite sua nota do 2º bimestre: "))
nota4 = float(input("Digite sua nota do 2º bimestre: "))
media = (nota1 + nota2 + nota3 + nota4) /4

# Processamento

if media >= 7:
    print(f"Parabéns {nome}, você foi APROVADO com a média de {media}. ")
elif media >= 5:
    print(f"{nome}, sua média foi de {media}, você precisará fazer RECUPERAÇÃO. ")
else:
  print(f"A sua média {nome}, foi de {media}, você foi REPROVADO!")
```

```python
# 4 - 4.	Classificação de idade
# Receba a idade e classifique:
# Criança (0–12)
# Adolescente (13–17)
# Adulto (18–59)
# Idoso (60+)

# Entrada

idade = int(input("Digite sua idade por favor: "))

# Processamento

if 0 <= idade <= 12:
  print(f"Você é uma criança!")
elif 13 <= idade <= 17:
  print(f"Você é um adolescente!")
elif 18 <= idade <= 59:
  print(f"Você é um adulto!")
else:
  print(f"Você está na melhor idade")
```

```python
# 5 - Senha correta
# Defina uma senha fixa no código. Peça ao usuário para digitar e informe se está correta ou incorreta.

# Entrada

senha_fixa = "QWERT123"

# Entrada
senha_do_usuario = input("Digite a senha por favor!")

# Processamento

if senha_fixa == senha_do_usuario:
    print("---------------------------------------------------------------------")
    print("Senha correta, acesso PERMITIDO!")
else:
  print("---------------------------------------------------------------------")
  print("Senha incorreta, acesso NEGADO!")
```

```python
# 6 - Positivo, negativo ou zero
# Peça um número e diga se ele é positivo, negativo ou zero.

# Entrada

num = float(input("Digite um número por favor: "))

# Processamento

if num > 0:
  print("O número é POSITIVO!")
elif num < 0:
  print("O número é NEGATIVO!")
else:
  print("O número é IGUAL A ZERO!")


```

```python
# 7 - Calculadora simples
# Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado.

# Entrada
num1 = float(input("Digite o primeiro número: "))
operacao = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

# Processamento e Saída
if operacao == '+':
  resultado = num1 + num2
  print(f"Resultado: {resultado}")
elif operacao == '-':
  resultado = num1 - num2
  print(f"Resultado: {resultado}")
elif operacao == '*':
  resultado = num1 * num2
  print(f"Resultado: {resultado}")
elif operacao == '/':
  if num2 != 0:
    resultado = num1 / num2
    print(f"Resultado: {resultado}")
  else:
    print("Erro: Divisão por zero não é permitida.")
else:
  print("Operação inválida.")
```

```python
# 8 - Dia da semana
# Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

# Entrada

num_dia = int(input("Digite um número entre 1 e 7." ))

# Processamneto e Saída

if num_dia == 1:
  print("É Domingo!")
elif num_dia == 2:
  print("É Segunda-feira")
elif num_dia == 3:
  print("É Terça-feira")
elif num_dia == 4:
  print("É Quarta-feira")
elif num_dia == 5:
  print("É Quinta-feira")
elif num_dia == 6:
  print("É Sexta-feira")
elif num_dia == 7:
  print("É Sábado!")
else:
  print("Dia inválido, digite um númeoro entre 1 e 7 por favor!")
```

```python
# 9.	Ano bissexto
# Solicite um ano e informe se ele é bissexto ou não.

# Entrada
ano = int(input("Digite um ano: "))

# Processamento e Saída
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
  print(f"O ano {ano} é bissexto!")
else:
  print(f"O ano {ano}, não é bissexto!")
```

```python
# 10 - Desconto em compras
# Peça o valor de uma compra:
# ≥ 100: 10% de desconto
# ≥ 50 e < 100: 5% de desconto
# < 50: sem desconto

# Entrada
valor_da_compra = float(input("Digite o valor da compra por favor: "))

# Processamento

if valor_da_compra >= 100:
  desconto = 0.10 # 10%
elif valor_da_compra >= 50:
  desconto = 0.05 # 5%
else:
  desconto = 0 # Sem desconto

valor_final = valor_da_compra * (1 - desconto)
print(f"O valor final da compra com desconto é: R${valor_final:.2f}")
```
