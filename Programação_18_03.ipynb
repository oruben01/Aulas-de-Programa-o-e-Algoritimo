```python
# Operadores arítiméticos

# Fazer um programa que o usuário digita 02 números.
# Saída mostrar operações.

# Entrada

a = int(input("Digite um número: "))
b = int(input("Digite um número: "))

# Processamento

soma = a + b
subtracao = a - b
multiplicação = a * b
divisão = a / b
div_int = a // b
resto = a % b
potencia = a ** b

# Saída

print("")
print("--------------------------OPERADORES ARÍTIMÉTICOS----------------------")
print("")
print("----------------------------------SOMA---------------------------------")
print("")
print(f"O resultado da SOMA de {a} MAIS {b} é igual a: {soma}")
print("-----------------------------------------------------------------------")
print("")
print("--------------------------OPERADORES ARÍTIMÉTICOS----------------------")
print("")
print("--------------------------------SUBTRAÇÃO------------------------------")
print("")
print(f"O resultado da SUBTRAÇÃO de {a} MENOS {b} é igual a: {subtracao}")
print("-----------------------------------------------------------------------")
print("")
print("--------------------------OPERADORES ARÍTIMÉTICOS----------------------")
print("")
print("------------------------------MULTIPLICAÇÃO----------------------------")
print("")
print(f"O resultado da MULTIPLICAÇÃO de {a} VEZES {b} é igual a: {multiplicação}")
print("-----------------------------------------------------------------------")
print("")
print("--------------------------OPERADORES ARÍTIMÉTICOS----------------------")
print("")
print("---------------------------------DIVISÃO-------------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} DIVIDIDO por {b} é igual a: {divisão}")
print("-----------------------------------------------------------------------")
print("--------------------------OPERADORES ARÍTIMÉTICOS----------------------")
print("")
print("------------------------------DIVISÃO INTEIRA--------------------------")
print("")
print(f"O resultado da DIVISÃO  INTEIRA de {a} DIVIDIDO por {b} é igual a: {div_int}")
print("-----------------------------------------------------------------------")
print("--------------------------OPERADORES ARÍTIMÉTICOS----------------------")
print("")
print("-------------------------------RESTO DIVISÃO---------------------------")
print("")
print(f"O RESTO da DIVISÃO de {a} DIVIDIDO por {b} é igual a: {resto}")
print("-----------------------------------------------------------------------")
print("--------------------------OPERADORES ARÍTIMÉTICOS----------------------")
print("")
print("-------------------------------POTENCIAÇÃO-----------------------------")
print("")
print(f"O resultado da POTENCIAÇÃO de {a} ELEVADO por {b} é igual a: {potencia}")
print("-----------------------------------------------------------------------")
```

```python
# Fazer um programa que converte km em metros.

# Entrada

km = int(input("Digite seu KM: "))

# Processamento

metros = km * 1000

# Saída

print(f"Seu KM é igual a: {metros}")
```
